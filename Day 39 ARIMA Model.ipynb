```python
# Drop if any null records

df.dropna(inplace=True)
display(df.head(50))
```

```python
df=pd.DataFrame(df['Temp'])
df.plot()
```

```python
# Plot the Trend and Seasonality

df['Temp'].rolling(30).mean().plot()
df['Temp'].rolling(30).std().plot()
df['Temp'].plot()
plt.show()
```

```python
# Seasonal Decompose

from statsmodels.tsa.seasonal import seasonal_decompose
Decomp_results=seasonal_decompose(df)
Decomp_results.plot()
plt.show()
```

```python
# Train Test Split

train=df.iloc[:510,0]
test=df.iloc[510:,0]
```

```python
# Create PACF Graph
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_pacf(train,lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_pacf(train,lags=5)
plt.show()
```

```python
plot_acf(train,lags=50)
plt.show()
```

```python
# Identify best parameters 

from pmdarima import auto_arima
auto_arima(df,trace=True)
```

```python
df
```

```python
# Create Model

from statsmodels.tsa.arima.model import ARIMA
A_Model=ARIMA(train,order=(1,1,2))
predictor=A_Model.fit()
predictor.summary() 
```

```python
df
```

```python
# Prdiction for Training data set

Predicted_results=predictor.predict(start=len(train),end=len(train)+len(test)-1,typ='levels')
display(Predicted_results)
```

```python
# Display Prediction and Test values

result=pd.concat([test,Predicted_results ],axis=1)
display(result)
```

```python
# Plot Graph

plt.plot(test,color='red',label='Actual Temp')
plt.plot(Predicted_results,color='blue',label='Predicted Temp')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.legend()
plt.show()
```

```python
# Caluculate the average

display(test.mean())
display(Predicted_results.mean())
```

```python
# Calculated RMSE

import math
from sklearn.metrics import mean_squared_error
rmse=math.sqrt(mean_squared_error(test,Predicted_results))
display(rmse)
```

```python
# Accuracy

from sklearn.metrics import r2_score
r2_score(test,Predicted_results)
```
